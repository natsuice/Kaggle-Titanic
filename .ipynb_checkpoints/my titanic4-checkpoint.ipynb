### 数据特征工程

```python
#填补缺失值
for data in total:
    data["Age"].fillna(data["Age"].median(), inplace=True)
    data["Embarked"].fillna(data["Embarked"].mode()[0], inplace=True)
    data["Fare"].fillna(data["Fare"].mode()[0], inplace=True)
    
    data.loc[train["Cabin"].notnull(), "Cabin"] = 1
    data["Cabin"].fillna(0, inplace=True)
```

```python
#删除不需要的特征
drop_feature = ["PassengerId", "Ticket"]

train.drop(drop_feature, 1, inplace=True)
```

```python
# 检查数据是否还存在缺失值
for data in total:
    print(data.isnull().sum()/len(data))
```

```python
#特征转换
for data in total:
    data["Title"] = data.Name.str.extract(" ([A-Za-z]+)\.", expand=False) #从Name中提取Title
    data["AgeBand"] = pd.cut(data["Age"], bins=10) #将Age划分等级
    data["FamilySize"] = data["SibSp"] + data["Parch"] + 1 #添加FamilySize特征
    data["IsAlone"] = 1 #根据FamilySize特征添加IsAlone特征
    data["IsAlone"].loc[data["FamilySize"] > 1] = 0 
    data["FareBand"] = pd.qcut(data["Fare"], 5) #将Fare划分等级
```

```python
#将Title细分
for data in total:
    data["Title"].replace(["Capt", "Col", "Don", "Dr", "Jonkheer", "Major", "Rev", "Sir"], "Rare", inplace=True)
    data["Title"].replace(["Countess", "Mme","Ms", "Dona"], "Mrs", inplace=True)
    data["Title"].replace(["Lady", "Mlle"], "Miss", inplace=True)
```

```python
#LabelEncode
le = LabelEncoder()
for data in total:
    data["Sex_Code"] = le.fit_transform(data["Sex"])
    data["Embarked_Code"] = le.fit_transform(data["Embarked"])
    data["Title_Code"] = le.fit_transform(data["Title"])
    data["AgeBand_Code"] = le.fit_transform(data["AgeBand"])
    data["FareBand_Code"] = le.fit_transform(data["FareBand"])
```

```python
#创造一些别的特征
for data in total:
    data["Pclass*Sex"] = data["Pclass"] * data["Sex_Code"]
    data["Embarked*Sex"] = data["Embarked_Code"] * data["Sex_Code"]
    data["Title*Sex"] = data["Title_Code"] * data["Sex_Code"]
    data["Title*Pclass"] = data["Title_Code"] * data["Pclass"]
    data["Title*IsAlone"] = data["Title_Code"] * data["IsAlone"]
    data["Title*FamilySize"] = data["Title_Code"] * data["FamilySize"]
    data["Title*Embarked"] = data["Title_Code"] * data["Embarked_Code"]
    data["Age*Sex"] = data["AgeBand_Code"] * data["Sex_Code"]
    data["Age*Pclass"] = data["AgeBand_Code"] * data["Pclass"]
    data["Age*Title"] = data["AgeBand_Code"] * data["Title_Code"]
    data["Age*FamilySize"] = data["AgeBand_Code"] * data["FamilySize"]
    data["Age*Embarked"] = data["AgeBand_Code"] * data["Embarked_Code"]
     
```

```python
#One hot encode
dummies = ["Sex", "Embarked", "Title"]

train_dummies = pd.get_dummies(train[dummies])
test_dummies = pd.get_dummies(test[dummies])

train = pd.concat([train, train_dummies], 1)
test = pd.concat([test, test_dummies], 1)
```

```python
#尝试不同的特征组合

#label
target = ["Survived"]

#所有数值型特征
data_num = ['Pclass', 'SibSp', 'Parch', 'FamilySize', 'IsAlone', 'Cabin',
            'Sex_Code', 'Embarked_Code', 'Title_Code', 'AgeBand_Code', 'FareBand_Code', 
            'Sex_female', 'Sex_male', 'Embarked_C', 'Embarked_Q', 'Embarked_S',
            'Title_Master', 'Title_Miss', 'Title_Mr', 'Title_Mrs', 'Title_Rare',
            'Age*Sex', 'Age*Pclass', 'Age*Title', 'Age*FamilySize', 'Age*Embarked',
            'Pclass*Sex', 'Embarked*Sex', 'Title*Sex', 'Title*Pclass', 'Title*IsAlone', 'Title*FamilySize', 'Title*Embarked']

#全是组合特征
data_new = ['Pclass', 'FamilySize', 'Sex_Code', 'Cabin',
            'Age*Sex', 'Age*Pclass', 'Age*Title', 'Age*FamilySize', 'Age*Embarked',
            'Pclass*Sex', 'Embarked*Sex', 'Title*Sex', 
            'Title*Pclass', 'Title*IsAlone', 'Title*FamilySize', 'Title*Embarked']

#将重复信息删除并不含新增特征的特征组合
data_x = ['Pclass', 'SibSp', 'Parch', 'FamilySize', 'IsAlone', 'Cabin',
          'Sex_Code', 'Embarked_Code', 'Title_Code', 'AgeBand_Code', 'FareBand_Code']

#表示范围的特征组合
data_bin = ['Pclass', 'FamilySize', 'Sex_Code', 'Cabin', 'Embarked_Code', 'Title_Code', 'AgeBand_Code', 'FareBand_Code']

#还有独热编码的特征组合
data_dummy = ['Pclass', 'FamilySize', 'Cabin', 'AgeBand_Code', 'FareBand_Code', 
                'Sex_female', 'Sex_male', 
                'Embarked_C', 'Embarked_Q', 'Embarked_S', 'Title_Master', 
                'Title_Miss', 'Title_Mr', 'Title_Mrs', 'Title_Rare']     

data_feature = {"data_num": data_num, 
                "data_x": data_x, 
                "data_new": data_new,
                "data_bin": data_bin, 
                "data_dummy":data_dummy}
```

### 建模并进行评估

```python
#选择DecisionTree
cv_split = ShuffleSplit(n_splits=10, test_size=0.3, train_size=0.6, random_state=0)

clf = [
       #决策树
       DecisionTreeClassifier(), 
    
       #逻辑回归
       LogisticRegression(),
    
       #AdaBoost
       AdaBoostClassifier(), 
    
       #Bagging
       BaggingClassifier(),
     
       #随机森林
       RandomForestClassifier(),
       
       #支持向量机
       SVC(),
      
       #K近邻
       KNeighborsClassifier(),
       
       #高斯贝叶斯
       GaussianNB()
       ]
```

```python
#创建一个模型与特征的组合选择表格
MLA_columns = ["MLA name", "Data", "Train Accuracy Mean", "Test Accuracy Mean"]
MLA_compare = pd.DataFrame(columns=MLA_columns)
row_index = 0

#交叉验证评估模型
for data_name, mla in product(data_feature, clf):
    data = data_feature[data_name]
    cv_result = cross_validate(mla, train[data], train[target], cv=cv_split)
    MLA_compare.loc[row_index, "MLA name"] = mla.__class__.__name__
    MLA_compare.loc[row_index, "Data"] = data_name
    MLA_compare.loc[row_index, "Train Accuracy Mean"] =  cv_result["train_score"].mean()*100
    MLA_compare.loc[row_index, "Test Accuracy Mean"] =  cv_result["test_score"].mean()*100
    row_index += 1
```

```python
clf = [
       #决策树
       DecisionTreeClassifier(), 
    
       #逻辑回归
       LogisticRegression(),
    
       #AdaBoost
       AdaBoostClassifier(), 
    
       #Bagging
       #BaggingClassifier(),

     
       #随机森林
       RandomForestClassifier(),
       
       #支持向量机
       #SVC(),
      
       #K近邻
       #KNeighborsClassifier(),
       
       #高斯贝叶斯
       #GaussianNB()
       ]
FS_columns = ["MLA name", "Data", "Before Train Mean", "Before Test Mean",
                     "After Data", "After Train Mean", "After Test Mean"]

FS = pd.DataFrame(columns=FS_columns)
row_index = 0

for data_name, mla in product(data_feature, clf):
    data = data_feature[data_name]
    cv_result = cross_validate(mla, train[data], train[target], cv=cv_split)
    FS.loc[row_index, "MLA name"] = mla.__class__.__name__
    FS.loc[row_index, "Data"] = data_name
    FS.loc[row_index, "Before Train Mean"] =  cv_result["train_score"].mean()*100
    FS.loc[row_index, "Before Test Mean"] =  cv_result["test_score"].mean()*100
    
    fs = RFECV(mla, step = 1, scoring = 'accuracy', cv = cv_split)
    fs.fit_transform(train[data], train[target])
    X_fs = train[data].columns.values[fs.get_support()]
    
    fs_cv_result = cross_validate(mla, train[X_fs], train[target], cv=cv_split)
    FS.loc[row_index, "After Data"] = X_fs
    FS.loc[row_index, "After Train Mean"] =  fs_cv_result["train_score"].mean()*100
    FS.loc[row_index, "After Test Mean"] =  fs_cv_result["test_score"].mean()*100
    
    row_index += 1

FS
```

```python
FS["++"] = FS["After Test Mean"] - FS["Before Test Mean"]
FS
```

```python
est = [
       #逻辑回归
       ("LGR", LogisticRegression()),
    
       #决策树
       ("DT", DecisionTreeClassifier()),
       
       #AdaBoost
       ("ada", AdaBoostClassifier()), 
    
       #Bagging
       ("bc", BaggingClassifier()),
     
       #随机森林
       ("rf", RandomForestClassifier()),
       
       #支持向量机
       ("svc", SVC()),
      
       #K近邻
       ("kn", KNeighborsClassifier()),
       
       #高斯贝叶斯
       ("gnb", GaussianNB())
       ]

param = [
        #LogisticRegression
        [{"solver": ["liblinear", "sag", "saga", "newton-cg", "lbfgs"],
          "warm_start": [True, False]}],
              
        #DecisionTree
        [{"criterion": ["gini", "entropy"],
          "max_depth": [2, 4, 6, 8, 10, 12, 14, 16, None],
          "min_impurity_decrease": np.linspace(0, 0.2, 50),
          "random_state": [0, None]}],
    
        #AdaBoostClassifier
        [{
            "n_estimators": [10, 50, 100, 300],
            "learning_rate": [0.1, 0.25, 0.5, 0.75, 1.],
            "algorithm": ["SAMME", "SAMME.R"],
            "random_state": [0, None]
        }],
    
        #BaggingClassifier
        [{
            "n_estimators": [10, 50, 100, 300],
            "max_samples": [0.1, 0.25, 0.5, 0.75, 1.],
            "random_state": [0, None]
        }],
    
        #RandomForestClassifier
        [{
            'n_estimators':  [10, 50, 100, 300],
            'criterion': ["gini", "entropy"],
            'max_depth': [2, 4, 6, 8, 10, 12, 14, 16, None],
            'oob_score': [True, False],
            'random_state': [0, None]
        }],
    
        #svc
        [{
            'C': [1,2,3,4,5], 
            'gamma':  [0.1, 0.25, 0.5, 0.75, 1.],
            'decision_function_shape': ['ovo', 'ovr'], 
            'probability': [True, False],
            'random_state': [0, None]
        }],
    
       #KNeighborsClassifier
        [{
            'n_neighbors': [1,2,3,4,5,6,7], 
            'weights': ['uniform', 'distance'], 
            'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
        }],
    
       #GaussianNB
        [{
            
        }]
         ]

TC_columns = ["MLA name", "Data", "Params", "Train score", "Test score"]

TC = pd.DataFrame(columns=TC_columns)
row_index = 0

for (clf, param), feature in product(zip(est, param), data_feature):
    tune_model = GridSearchCV(clf[1], param_grid=param, scoring="roc_auc", cv=cv_split)
    tune_model.fit(train[data_feature[feature]], train[target])
    #print(feature)
    TC.loc[row_index, "Data"] = feature
    #print(clf[1].__class__.__name__)
    TC.loc[row_index, "MLA name"] = clf[1].__class__.__name__
    #print("After tune parameters:")
    TC.loc[row_index, "Params"] = [tune_model.best_params_]
    #print(tune_model.best_params_)
    #print("test score:{:.2f}".format(tune_model.cv_results_["mean_test_score"][tune_model.best_index_]*100))
    TC.loc[row_index, "Train score"] = tune_model.cv_results_["mean_train_score"][tune_model.best_index_]*100
    TC.loc[row_index, "Test score"] = tune_model.cv_results_["mean_test_score"][tune_model.best_index_]*100
    row_index += 1
    
```

```python
#Confusion Matrix函数
def plot_confusion_matrix(cm, classes, 
                          normalize = False, 
                          title="Confusion Matrix",
                          cmap=plt.cm.Blues):
    
    if normalize:
        cm = cm.astype("float")/cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print("Confusion matrix without normalization")
    
    print(cm)
    
    plt.imshow(cm, interpolation="nearest",cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    fmt = ".2f" if normalize else "d"
    thresh = cm.max()/2.0
    for i,j in product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i,j] > thresh else "black")
        
    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.tight_layout()      
```

```python
#查看模型confusion matrix
cm = confusion_matrix(train[target], y_pred)
classes = ["Dead", "Survived"]

print(classification_report(train[target], y_pred))

plt.figure()
plot_confusion_matrix(cm=cm, classes=classes)

plt.figure()
plot_confusion_matrix(cm=cm, classes=classes, title="Confusion Matrix with noramlization", normalize=True)
```
